14. how is memory managed in python?
Ans: In Python, memory management is handled automatically by the Python interpreter. Python uses a technique called "reference counting" to keep track of objects in memory and determine when they can be deleted. When an object is created in Python, the interpreter assigns it a unique ID and a reference count of one. The reference count is incremented each time the object is referenced or stored in another variable.

When an object's reference count drops to zero, Python automatically deletes the object and frees up its memory. This process is known as "garbage collection". Python also uses a separate mechanism called "cyclic garbage collection" to detect and remove circular references, where two or more objects reference each other in a cycle and have a reference count of one.
